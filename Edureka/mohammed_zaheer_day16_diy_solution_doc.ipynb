**Q1. Problem Statement: Sales Data Analysis- I**

Load the “sample_superstore.csv” dataset provided on LMS into a DataFrame, using the Matplotlib library, plot the following insights from the dataset:

1. Compare the total sales for each category of products
2. Number of deals in each state of America

**Note:** You can use any plot of your choice to show the insights

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("SampleSuperstore.csv")
#display(df)

#1.Compare the total sales for each category of products:

#Applying bar chart matplotlib:
#--> syntax: plt.bar(Column for X - axis,Column for Y - axis)

plt.bar(df.Category, df.Sales)

#setting labels for X and Y axes:

plt.xlabel("Category")
plt.ylabel("Sales")

#Setting title for the bar chart
plt.title("Total Sales for each Category of Products")

#displaying bar chart output:

plt.show()

#2. Number of deals in each state of America:

df.State.value_counts().plot(kind='bar',width=0.8)
plt.ylabel("Deals")
plt.title("Total Number of Deals in Each American State")
```

**Q2. Problem Statement: Data Grouping - II**

Load the “sample_superstore.csv” dataset provided on LMS into a DataFrame; using the Seaborn library, plot the following insights from the dataset:

1. Print a correlation matrix between Postal code, Quantity, Sales, Discount,and Profit

2. Plot the correlation matrix using a heatmap

**Note:** You can use any plot of your choice to show the insights

```python
import pandas as pd
import seaborn as sb

#reading the file

data=pd.read_csv("SampleSuperstore.csv")
#print(data.corr())

#applying heatmap correlation of seaborn

dataplot = sb.heatmap(data.corr(), annot=True)
plt.show(dataplot)
```
