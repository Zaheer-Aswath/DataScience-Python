**Q1. Problem Statement: Working with DataFrames - I**

Write a Python program that reads the “iris.csv” dataset (Available on LMS) into a DataFrame, and filters the rows on the condition that “sepal length” should be more than or equal to 6cm. Write the filtered rows in a new DataFrame, and save the new DataFrame as a CSV file.

```python
import pandas as pd
df = pd.read_csv('Iris.csv')
#df

#filter_condition is a variable which stores conditions required from dataframe(df) of iris.csv file
filter_condition=df["SepalLengthCm"] >= 6
print("Flowers with morethan 6 cm sepal length are:")
df[filter_condition] #applying stored condition in dataframe(df)
```

```python
#saving result to new csv(missed to upload in edureka)

df[filter_condition].to_csv("filtered_iris.csv")
```
