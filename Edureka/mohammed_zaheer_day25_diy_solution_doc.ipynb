Q1. Problem Statement: Probability Distribution

Load the exam.csv dataset into a DataFrame and perform the following tasks on the “math score,” ”reading score,” and “writing score” columns:

1. Plot the distributions using the seaborn library
2. Find the mean and standard deviation using the scipy library
3. Find skewness and kurtosis using the stats module from the scipy library
4. Find Standard Normal Distribution (z-score) using the stats module from the scipy library
5. Plot the Z distribution using the seaborn library

```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy import mean
from scipy import std
import warnings # using this remove depricated warning message
warnings.filterwarnings('ignore',message='.*is deprecated and will be removed in SciPy 2.0.0*.')#filtered warning message

exam_df = pd.read_csv("exam.csv")
#display(exam_df)

##1. Plot the distributions using the seaborn library:

print("1. A distribution plot for given columns:\n")
fig, axes = plt.subplots(1,3, figsize=(15,5))

dist_math=sb.kdeplot(exam_df, ax=axes[0], x="math score")
dist_read=sb.kdeplot(exam_df,ax=axes[1], x="reading score")
dist_writ=sb.kdeplot(exam_df,ax=axes[2], x="writing score")
plt.show(dist_math)
plt.show(dist_read)
plt.show(dist_writ)

'''#sb.displot(exam_df, x="math score")
#displot is deprecated hence using kdeplot
dist_math = sb.displot(exam_df, x="math score",kind="kde")
plt.show(dist_math)

dist_read = sb.displot(exam_df, x="reading score",kind="kde")
plt.show(dist_read)

dist_writ = sb.displot(exam_df, x="writing score",kind="kde")
plt.show(dist_writ)'''

##2. Mean and Standard deviation for a given column(math score,reading score,writing score) using the scipy library:

print("\n2. Mean and Standard deviation for a given column(math score,reading score,writing score):\n")

mean_math = scipy.mean(exam_df["math score"])
print("\nMean of math score:",mean_math)
std_math = scipy.std(exam_df["math score"])
print("Standard deviation of math score:",std_math)

mean_read = scipy.mean(exam_df["reading score"])
print("\nMean of reading score:",mean_read)
std_read = scipy.std(exam_df["reading score"])
print("Standard deviation of reading score:",std_read)

mean_writ = scipy.mean(exam_df["writing score"])
print("\nMean of writing score:",mean_writ)
std_writ = scipy.std(exam_df["writing score"])
print("Standard deviation of writing score:",std_writ)

##3. Skewness and kurtosis for a given column (math score,reading score,writing score) using the stats module from the scipy library:

print("\n3. Skewness and kurtosis for a given column (math score,reading score,writing score):\n")

math_skew = stats.skew(exam_df["math score"])
print("\nThe Skewness of the math score:",math_skew)
math_kurt = stats.kurtosis(exam_df["math score"])
print("The Skewness of the math score:",math_kurt)

read_skew = stats.skew(exam_df["reading score"])
print("\nThe Skewness of the reading score:",read_skew)
read_kurt = stats.kurtosis(exam_df["reading score"])
print("The Skewness of the reading score:",read_kurt)

writ_skew = stats.skew(exam_df["writing score"])
print("\nThe Skewness of the writing score:",writ_skew)
writ_kurt = stats.kurtosis(exam_df["writing score"])
print("The Skewness of the writing score:",writ_kurt)

##4. Standard Normal Distribution (z-score) for a given column (math score,reading score,writing score) using the stats module from the scipy library:

print("\n4. Standard Normal Distribution (z-score) for a given column (math score,reading score,writing score):\n")

math_zscore = stats.zscore(exam_df["math score"])
print("\nZscore of the math score:",np.array(math_zscore)) #answer requested in array

read_zscore = stats.zscore(exam_df["reading score"])
print("\nZscore of the reading score:",np.array(read_zscore))

writ_zscore= stats.zscore(exam_df["writing score"])
print("\nZscore of the writing score:",np.array(writ_zscore))

##5. Plot for Z-distribution using the seaborn library

print("\n5. Plot for Z-distribution\n")
fig, axes = plt.subplots(1,3, figsize=(15,5))

z_dist_math=sb.kdeplot(exam_df, ax=axes[0], x=math_zscore)
z_dist_read=sb.kdeplot(exam_df,ax=axes[1], x=read_zscore)
z_dist_writ=sb.kdeplot(exam_df,ax=axes[2], x=writ_zscore)
plt.show(z_dist_math)
plt.show(z_dist_read)
plt.show(z_dist_writ)
```
