**Q1. Problem Statement: Measures of Central Tendency- I**

Write a Python program to declare a 1-D array. With the help of the statistics library, calculate the Mean, Median, and Mode.

**Note:** You can write any data of your choice inside the array

```python
import pandas as pd
import numpy as np
from statistics import mean
from statistics import median
from statistics import mode

x=[1,2,3,4,5,6,2,3,4,5,5,5,5]
y=np.array(x)

print("Mean of the Array is: \n", mean(y))
print("Median of the Array is: \n", median(y))
print("Mode of the Array is: \n", mode(y))
```

**Q2. Problem Statement: Measures of Central Tendency- II**

Write a Python program to declare a DataFrame as shown in the dataset section below and calculate the Mean, Median, and Mode of the entire DataFrame.

**Note:** You can write any data of your choice inside the DataFrame

```python
import pandas as pd
import numpy as np
from statistics import mean
from statistics import median
from statistics import mode

# initialize data of lists.
data = {'Apple': [10, 7, 55, 15,7,5],'Orange': [20,14,15,14,1,4],'Banana': [30,21,8,1,1,9],'Pear': [40,28,12,8,8,2]}

# Creates pandas DataFrame.
df = pd.DataFrame(data, index=['Bag1','Bag2','Bag3','Bag4','Bag5','Bag6'])

display(df)

print("\nCalculating the Mean\n")
print(df.mean())
print("\nCalculating the Median\n")
print(df.median())
print("\nCalculating the Mode\n")
print(df.mode())
```

**Q3. Problem Statement: Measures of Central Tendency- III**

**Load the “Books.csv” dataset into a DataFrame and perform the following tasks:**

1. Perform initial analysis on the dataset using the info() and the describe() functions

2. Group the data based on the "Author" column and calculate the average user ratings and average price, and write them in new columns

3. Find out which of the authors have written the maximum number of books

```python
from pandas.core.groupby import groupby
import pandas as pd
import numpy as np

df1 = pd.read_csv("Books.csv")
#display(df1)

##Case - 1:

print("\n1. Perform initial analysis of the dataset using the info() and the describe() functions\n")
df1.info()
print("\n",df1.describe())

##Case - 2:

print("\n2. Group the data based on the Author column and calculate the average user ratings and average price, and write them in new columns\n")

#Group the data based on the "Author" column and calculate the average user ratings and write them in new columns

author_avguser = df1.groupby(["Author"],as_index=False).mean("User Rating")
author_avguser.drop(columns = {"Reviews","Price","Year"},inplace=True)
author_avguser.rename(columns = {'User Rating':'Average_User_Rating'}, inplace = True)
display(author_avguser)

#Group the data based on the "Author" column and calculate the average price and write them in new columns

author_avgprice = df1.groupby(["Author"],as_index=False).mean("Price")
author_avgprice.drop(columns = {"User Rating","Reviews","Year"},inplace=True)
author_avgprice.rename(columns = {'Price':'Average_Price'}, inplace = True)
display(author_avgprice)

##Case - 3:

print("\n3. Find out which of the authors have written the maximum number of books\n")
print("\nMaximum number of books written by:")
print(df1["Author"].mode())
```
