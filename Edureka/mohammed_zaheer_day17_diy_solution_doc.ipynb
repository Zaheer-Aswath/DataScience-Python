**Q1. Problem Statement: Data Visualization using Matplotlib**

Load the “Netflix_titles.csv” data into a DataFrame and plot the following insights using the Matplotlib library:

1. How much of the Netflix content are TV shows or Movies?

2. In which country is Netflix the most popular?

```python
##Step-1: importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

##Step-2: reading the file

net=pd.read_csv("netflix_titles.csv") #net is variable stores DataFrame
#display(net)

###Case-1: How much of the Netflix content are TV shows or Movies?

##Step-3: Filtering Movie and TV Show from column "type" of "netflix_titles.csv" and counting no.of Movie and TV Show

net_type=set(net["type"])  #net_type stores unique data from "type"
#print(net_type)

##Step-4: Here we using Dictionaries to count and store data
   #--> net_dict is variable to store keys(Movie and TV show) and values(count) from "type"

net_dict={}

##Step-5: using for loop for counting  - Movie and TV show from "type"

for i in net_type:
  count=len(net[net["type"]==i])
  net_dict[i]=count

##Step-6: Checking the keys and values from net_dict after looping

#print(net_dict)

##Step-7: Plotting the data to pie chart

print("1. How much of the Netflix content are TV shows or Movies?\n")
plt.figure(figsize=(10,8))
myexplode = [0,0.1] #myexplode is variable contains list[0,0.1] as per net_dict["Movie","TV Show"]
plt.pie(net_dict.values(), labels=net_dict.keys(),autopct='%.2f%%',explode=myexplode)
plt.title("Type of Netflix Content")
plt.show()

###Case-2: In which country is Netflix the most popular

##Step-1: Counting(Nos:748) the countries name and resetting index (counting countries only when it is present without comma)

net=net["country"].value_counts().reset_index()

##Step-2: filtering - For each country count, divide it by total count sum to get percentage of that country
      ##and filter records for which percentage is more than 1%

net=net[net["country"]/net["country"].sum()>0.01]

##Step - 3: Plotting the data to pie chart

print("2. In which country is Netflix the most popular?\n")
plt.figure(figsize=(10,8))
labels=net["index"] # index has countries
plt.pie(net["country"], labels=labels)
plt.title("Netflix Viewers across the Globe")
plt.show()
```

**Q2. Problem Statement: Data Visualization using Seaborn**

Load the “starbucks_drinkMenu_expanded.csv” data into a DataFrame and find out which “Beverage_category” has the highest calories; plot the insights using a bar plot with the help of the Seaborn library.

```python
##Step-1: importing libraries

import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

##Step-2: reading the file

star=pd.read_csv("starbucks_drinkMenu_expanded.csv") #star is variable stores DataFrame
#display(star)

##Step-3: checking if any column have null values

star.isnull().sum()

##Step-4: dropping the null values

star.dropna(inplace=True)

##Step-5: again checking whether any column have null values

star.isnull().sum()

##Step-6: Plotting the data to bar in seaborn

plt.figure(figsize=(10,8))
sb.barplot(x="Calories",y="Beverage_category",data=star).set(title="Which of the starbucks drinks have highest calories?")
plt.show()
```
