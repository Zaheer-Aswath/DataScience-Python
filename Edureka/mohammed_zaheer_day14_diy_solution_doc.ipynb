**Q1. Problem Statement: Handling Null Values**

Load the “titanic.csv” file provided on LMS into a DataFrame, find the columns with null values, and drop all the rows in DataFrame with null values to create a new DataFrame.

```python
import pandas as pd

df = pd.read_csv("titanic.csv")
display(df)

#columnwise null values from titanic.csv:

print("\n1. Identifying null values:\n")
display(df.isnull().sum())

#dropping the null values from titanic.csv:

print("\n2. Dropping the null values:\n")
df1 = df.dropna(axis = 0)
display(df1)

#Verifying there is no null values from titanic.csv of df1:

print("\n3. Verifying there are no null values in the new DataFrame:\n")
df2 = df1.isna()
display(df2)
```

**Q2. Problem Statement: Handling Duplicate Values**

Load the “titanic.csv” file provided on LMS into a DataFrame, find out the fields in the dataset with duplicate values based on the “Ticket” column, and drop all the rows with duplicate values to create a new DataFrame.

```python
import pandas as pd

df = pd.read_csv("titanic.csv")
display(df)

#removing or droping duplicated values from column "Ticket"

df3 = df.drop_duplicates(subset = 'Ticket')
display(df3)
```
